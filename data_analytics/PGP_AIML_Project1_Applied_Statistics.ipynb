### Approach / Requirements
The goal is to implement the Bayesian form of Logistic Regression. This is done by setting up a model relating the inputs (features) to the outputs (response) and then sampling from the posteriors for the model parameters. In the case of logistic regression, the model parameters are the weights on the features. The end result is a distribution (trace) of all the model parameter values that are drawn from the posterior. The procedure will be implemented in the Python library PyMC3, which is used for Bayesian inference and has a number of Markov Chain Monte Carlo implementations for sampling from the posterior. ### Helper Functions

```python
# Function to calculate missing values by column
def missing_values_table(df):
        # Name of the supplied dataframe
        name =[x for x in globals() if globals()[x] is df][0]
        
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print("Provided dataframe is: {0}\n\
The provided dataframe has {1} columns.\n\
There are {2} columns that have missing values.".format(
            name,
            str(df.shape[1]),
            str(mis_val_table_ren_columns.shape[0])
        ))
        
        # Return the dataframe with missing information
        if mis_val_table_ren_columns.shape[0]:
            return mis_val_table_ren_columns
        else:
            return None
```

## Interpret Statistical Evidence
### Do charges of people who smoke differ significantly from the people who don't?
https://www.kaggle.com/itsimplified/hypothesis-testing-insurance-claim


```python
#Null Hypothesis--> H0 = "charges has no effect on smoking"
#Alternate hypothesis--> H1 = "charges has effect on smoking" 

x = np.array(insurance_dataset[insurance_dataset['smoker'] == '1']['charges'])#selecting charges values corresponding to smoker as an array
y = np.array(insurance_dataset[insurance_dataset['smoker'] == '2']['charges'])#selecting charges values corresponding to not smoker as an array
t,p_value = stats.ttest_ind(x,y,axis =0)#performing an independent T-test
if p_value <0.05:
    print("charges has effect on smoking(reject H0)")
else:
    print("charges has no effect on smoking(accept H0)")
```
